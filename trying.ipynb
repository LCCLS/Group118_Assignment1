```python
import csv
import statistics
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
document_name = "Group118_Assignment1/ODI-2022.csv"
df = pd.read_csv(document_name, delimiter=";")
```

```python
df.describe()
```

```python
df.rename(columns={df.columns[0]: 'timestamp'},inplace=True)
df.rename(columns={df.columns[1]: 'program'},inplace=True)
df.rename(columns={df.columns[2]: 'course_on_ML'},inplace=True)
df.rename(columns={df.columns[3]: 'course_on_IR'},inplace=True)
df.rename(columns={df.columns[4]: 'course_on_stat'},inplace=True)
df.rename(columns={df.columns[5]: 'course_on_databases'},inplace=True)
df.rename(columns={df.columns[6]: 'gender'},inplace=True)
df.rename(columns={df.columns[7]: 'chocolate_makes_you'},inplace=True)
df.rename(columns={df.columns[8]: 'birthday'},inplace=True)
df.rename(columns={df.columns[9]: 'nr_neighbor'},inplace=True)
df.rename(columns={df.columns[10]: 'stand_up'},inplace=True)
df.rename(columns={df.columns[11]: 'stress_lev'},inplace=True)
df.rename(columns={df.columns[13]: 'rand_num'},inplace=True)
df.rename(columns={df.columns[14]: 'bed_time_yesterday'},inplace=True)
df.rename(columns={df.columns[15]: 'good_day1'},inplace=True)
df.rename(columns={df.columns[16]: 'good_day2'},inplace=True)
df.head()
```

```python
# Looking at most common program responses
print(df["program"].value_counts())
```

```python
# Preprocessing program variable.
# Compiled answers for the same program under one label for the top 3 most frequent answers
# Answers with less than 3 responses got categorized as "other"
df["program"] = ["AI" if (x.lower() == "ai" or "artificial" in x.lower()) else x for x in df["program"]]
df["program"] = ["CS" if ("computer" in x.lower() or "computational" in x.lower()) else x for x in df["program"]]
df["program"] = ["BA" if ("business" in x.lower()) else x for x in df["program"]]
top3 = ["AI", "BA", "CS"]
df["program"] = ["Other" if ((x not in top3) and (len(df[df.program == x]) < 3)) else x for x in df["program"]]
```

```python
df["program"].value_counts() # Most frequent responses after pre-processing
```

```python
to_be_cleaned = ["course_on_ML", "course_on_stat", "course_on_databases"]
affirm = ["yes", "ja", "mu"]
deny = ["no", "sigma", "nee"]
for col in to_be_cleaned:
    for x in df[col]:
        if x in affirm:
            df[col] = df[col].replace(x,1)
        elif x in deny:
            df[col] = df[col].replace(x,0)
```

```python
# Replacing all numerical values in the good_day columns with missing values
clean_up = ["good_day1", "good_day2"]
for col in clean_up:
    for x in df[col]:
        if x.isnumeric() == True:
            df[col] = df[col].replace(x, np.nan)
```

```python
# Cleaning up stress level variable
df["stress_lev"] = [np.nan if (str(x).isnumeric() == False or int(x) > 100 or int(x) < 0) else x for x in df["stress_lev"]]
```

```python
# Cleaning up neighbors variable
df["nr_neighbor"] = [np.nan if (str(x).isnumeric() == False or int(x) > 15 or int(x) < 0) else x for x in df["nr_neighbor"]]
```
